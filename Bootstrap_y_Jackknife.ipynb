# Bootstrap

El Bootstrap es una técnica de remuestreo no paramétrica que permite aproximar la distribución muestral de casi cualquier estadístico (como la media, la varianza o los intervalos de confianza) a partir de una única muestra original, sin hacer suposiciones estrictas sobre la distribución teórica de la población subyacente.

El método funciona asumiendo que la muestra original representa adecuadamente a la población. A partir de ella, se generan múltiples "pseudo-muestras" extrayendo datos al azar **con reemplazo**.

**Procedimiento:**
1. Partimos de una muestra original de tamaño $n$: $X = \{x_1, x_2, \dots, x_n\}$.
2. Seleccionamos aleatoriamente, y **con reemplazo**, $n$ elementos de $X$ para formar una nueva muestra Bootstrap $X^*_b$.
3. Calculamos el estadístico de interés $\hat{\theta}^*_b$ para esta nueva muestra.
4. Repetimos los pasos 2 y 3 un número grande de veces ($B$ iteraciones, por ejemplo, $B = 1000$).
5. Utilizamos la distribución empírica de los $B$ estadísticos calculados para estimar el error estándar o construir intervalos de confianza mediante percentiles.

```python
import numpy as np
```

Muestra de 12 observaciones y sacamos la media

```python
muestra = np.array([35, 42, 38, 40, 45, 37, 39, 41, 44, 36, 43, 40])
n = len(muestra)
print(f"Media original: {np.mean(muestra):.4f}")
```

Generar 1000 muestras Bootstrap

```python
B = 1000
medias_bootstrap = np.zeros(B)

np.random.seed(42) # Semilla para que los resultados sean reproducibles
```

Tenemos que es un muestreo con reemplazo y calculamos la media para cada B

```python
for i in range(B):

    muestra_b = np.random.choice(muestra, size=n, replace=True)

    medias_bootstrap[i] = np.mean(muestra_b)
```

Para los cálculos finales: obtenemos la media de las 1000 muestras, calculamos la desviación estándar y determinaos los percentiles 2.5% y 97.5% para construir el intervalo de confianza del 95%

```python
media_boot = np.mean(medias_bootstrap)

std_boot = np.std(medias_bootstrap, ddof=1)

ic_inferior = np.percentile(medias_bootstrap, 2.5)
ic_superior = np.percentile(medias_bootstrap, 97.5)
```

```python
print(f"Media de las 1000 muestras: {media_boot:.4f}")
print(f"Desviación estándar: {std_boot:.4f}")
print(f"Intervalo de confianza 95%: [{ic_inferior:.4f}, {ic_superior:.4f}]")
```

# Jackknife

El Jackknife es una técnica de remuestreo sistemática utilizada principalmente para estimar el sesgo y la varianza de un estimador estadístico. A diferencia del Bootstrap, que utiliza muestreo aleatorio con reemplazo, el Jackknife es determinista y consiste en recalcular el estadístico dejando fuera una observación a la vez.

**Procedimiento:**
1. Partimos de una muestra original de tamaño $n$: $X = \{x_1, x_2, \dots, x_n\}$.
2. Para cada observación $i$ (desde $1$ hasta $n$), creamos una submuestra de tamaño $n-1$ eliminando temporalmente el dato $x_i$.
3. Calculamos el estadístico de interés $\hat{\theta}_{(i)}$ para esta submuestra.
4. Obtenemos el promedio del estadístico Jackknife:
   $$\hat{\theta}_{jack} = \frac{1}{n} \sum_{i=1}^{n} \hat{\theta}_{(i)}$$
5. Estimamos la varianza del estimador utilizando la fórmula:
   $$\text{Var}_{jack} = \frac{n-1}{n} \sum_{i=1}^{n} \left( \hat{\theta} - \hat{\theta}_{(i)} \right)^2$$

Muestra de 12 observaciones y sacamos la media

```python
muestra = np.array([35, 42, 38, 40, 45, 37, 39, 41, 44, 36, 43, 40])
n = len(muestra)

print(f"Media original: {np.mean(muestra):.4f}\n")
```

Creamos una lista vacía para guardar las medias

```python
medias_jackknife = []
```

Eliminamos uno de los datos para obtener n muestras de tamaño n-1

```python
for i in range(n):
    # np.delete quita el elemento en la posición 'i'
    muestra_j = np.delete(muestra, i)

    # Calculamos el estadístico para cada muestra
    medias_jackknife.append(np.mean(muestra_j))
```

Obtenemos el promedio del Jackknife y calculamos la varianza

```python
theta_jack = np.mean(medias_jackknife)

suma_diferencias_cuadradas = np.sum((theta_jack - medias_jackknife)**2)
var_jack = ((n - 1) / n) * suma_diferencias_cuadradas
```

```python
print(f"Promedio del Jackknife: {theta_jack:.4f}")
print(f"Varianza del Jackknife: {var_jack:.4f}")
```
