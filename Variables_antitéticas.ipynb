## Variables antitéticas

El método de variables antitéticas es una técnica utilizada en la simulación de Monte Carlo para reducir la varianza de un estimador sin tener que aumentar el tamaño de la muestra ($N$), disminuyendo así el costo computacional.

El principio matemático se basa en inducir una covarianza negativa entre dos variables aleatorias. Si generamos un valor aleatorio $U$ distribuido uniformemente en $[0, 1]$, el valor $1 - U$ también sigue una distribución uniforme en el mismo intervalo. Al utilizar ambos valores, las fluctuaciones aleatorias tienden a compensarse.

**Procedimiento para estimar una integral $\int_0^1 f(x)dx$:**
1. Generar $N/2$ números pseudoaleatorios $U_i \sim U(0,1)$.
2. Evaluar la función con la variable original: $X_1 = f(U_i)$.
3. Evaluar la función con la variable antitética: $X_2 = f(1 - U_i)$.
4. Calcular el estimador antitético para cada par: $Y_i = \frac{f(U_i) + f(1 - U_i)}{2}$.
5. El estimador final $\hat{\theta}$ es el promedio de los $Y_i$, el cual tendrá una varianza menor que el estimador de Monte Carlo crudo.

```python
import random as r
import numpy as np
```

Definimos la función que quieremos integrar: exp(x^2)

```python
def f(x):
    return np.exp(x**2)
```

Monte Carlo Crudo

```python
def mc1(f, N):
    s = 0
    for i in range(N):
        U = r.random()
        s += f(U)
    return s / N
```

Prueba Monte Carlo Crudo

```python
l_crudo = []
# Hacemos 100 estimaciones, cada una con una muestra de 1000
for i in range(100):
    M = mc1(f, 1000)
    l_crudo.append(M)

print("Media:", np.mean(l_crudo))
print("Varianza:", np.var(l_crudo))
```

Variables Antitéticas

```python
def mc(f, N):
    s = 0
    for i in range(N):
        U = r.random()
        # Se evalúa U y 1-U, y se saca el promedio
        s += (f(U) + f(1 - U)) / 2
    return s / N
```

Prueba Variables Antitéticas

```python
l_anti = []
# Hacemos 100 estimaciones, cada una con una muestra de 1000
for i in range(100):
    M = mc(f, 1000)
    l_anti.append(M)

print("Media:", np.mean(l_anti))
print("Varianza:", np.var(l_anti))
```
