# Prueba de los promedios

La prueba de los promedios es una prueba estadistica que busca determinar si el conjunto de n numeros (100 en este caso) aleatorios tinen un promedio de $0.5$. se utiliza una distribuciòn normal para validar la hipotesis nula $H_0$


Las hipotesis son: //
  $H_0 = 0.5 $ //
  $H_a = 0.5 $

```python
import numpy as np
import scipy.stats as st
```

Hacemos una función para que calcule la media, cuente los elementos de la lista, lo estandarizamos y sacamos e, valor critico

```python
def promedios (data, alfa):
  x = np.mean(data)
  N = len(data)
  z = (x -0.5) * np.sqrt(N)/np.sqrt(1/12)
  z_c = st.norm.ppf(1 - alfa/2)
  if abs(z) < z_c:
    print("No se rechaza la hipotesis nula: La muestra proviene de una distribución U(0, 1)")
  else:
    print("Se rechaza la hipotesis nula: La muestra no proviene de una distribución U(0, 1)")
  return x, N
```

Definimos el intervalo de confianza e imprimimos el resultado

```python
media, lon = promedios(data, 0.05)
print("La media es: ", media)
```

Scamos el numero de frecuencias observadas $o_i$ en cada intervalo

```python
conteo1 = 0
conteo2 = 0
conteo3 = 0
conteo4 = 0
conteo5 = 0

for x in data:
  if 0 <= x < 0.2:
    conteo1 += 1
  elif x >= 0.2 and x < 0.4:
    conteo2 += 1
  elif 0.4 <= x < 0.6:
    conteo3 += 1
  elif 0.6 <= x < 0.8:
    conteo4 += 1
  elif 0.8 <= x <= 1:
    conteo5 += 1
conteo1, conteo2, conteo3, conteo4, conteo5
```

Sacamos las frecuencias esperadas $e_i = np$

```python
P = 1 / 5
N = len(data)
valor_esperado = N * P

# Crea una lista con el mismo valor repetido 5 veces
esperados = [valor_esperado] * 5
conteos = [conteo1, conteo2, conteo3, conteo4, conteo5]
esperados, conteos
```

Calculamos el estadistico $X^2$

```python
chi_cuadrado = 0
for i in range(5):
  chi_cuadrado += (conteos[i] - esperados[i])**2 / esperados[i]

print(f"El valor de X² es: {chi_cuadrado}")
```
