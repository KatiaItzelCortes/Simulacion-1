#Generación de Variables Aleatorias Discretas
Simular una variable tal que $p_1 = 0.20$, $p_2 = 0.15$, $p_3 = 0.25$, $p_4 = 0.40$, donde $p_j = P\{x \geq j\}$

```python
import numpy as np
import random as r
import matplotlib.pyplot as plt
```

```python
def f(n):
    Lista =[]
    for i in range(n):
        R = r.random()
        if 0 <= R <= 0.2:
            X = 1
        elif 0.2 < R <= 0.35:
            X = 2
        elif 0.35 < R <= 0.6:
            X = 3
        else:
            X = 4
        Lista.append(X)
    return Lista
```

```python
M = f(1000)
```

```python
plt.hist(M, density=True)
```
