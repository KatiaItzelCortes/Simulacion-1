## Prueba de racha

El objetivo primordial de esta prueba es determinar si una secuencia de números aleatorios presenta algún tipo de patrón o tendencia
En nuestro caso, trabajamos con rachas ascendentes y descendentes.

Si $x_{i+1} \geq x_i$, asignamos un signo (+) (Racha ascendente).

Si $x_{i+1} < x_i$, asignamos un signo (-) (Racha descendente).

Establecemos las siguientes hipótesis para un nivel de confianza del 95% ($\alpha = 0.05$):

$H_0$: Los números de la secuencia son independientes (las rachas ocurren de forma aleatoria).

$H_a$: Los números de la secuencia no son independientes (existe una tendencia o dependencia).

El número total de rachas ($co$) tiende a seguir una distribución normal con los siguientes parámetros:

Media Esperada ($\mu_{co}$) $$\mu_{co} = \frac{2n - 1}{3}$$
Varianza ($\sigma^2_{co}$) $$\sigma^2_{co} = \frac{16n - 29}{90}$$
Estadístico de Prueba ($Z_0$) $$Z_0 = \frac{a - \mu_a}{\sigma_a}$$

Criterio de Aceptación para un nivel de significancia de $\alpha = 0.05$, el valor crítico de la distribución normal es $Z_{\alpha/2} = 1.96$.

Si $|Z_0| \leq 1.96$:No se rechaza $H_0$. Se concluye que los números son independientes.

Si $|Z_0| > 1.96$: Se rechaza $H_0$. Los números presentan una estructura no aleatoria.

Lo primero que hace es cambiar la lista en una lista de signos binarios.

Una vez que tiene la lista de ceros y unos, el código debe identificar cuántos grupos hay.

Aplicamos las formulas para la media y la varianza y sacamos es estadistico y vemos si se recgaza o no la hipotesis nula.

```python
def prueba_rachas(secuencia):
    n = len(secuencia)
    N = n - 1

    signos = []
    for i in range(N):
        if secuencia[i+1] >= secuencia[i]:
            signos.append(1)
        else:
            signos.append(0)

    rachas = 1 # Vamos a contar el numero de rachas que tenemos
    for i in range(len(signos) - 1):
        if signos[i] != signos[i+1]:
            rachas += 1

    media_co = (2 * N - 1) / 3
    varianza_co = (16 * N - 29) / 90
    z_obs = (rachas - media_co) / np.sqrt(varianza_co)

    return rachas, media_co, z_obs

co, mu, z = prueba_rachas(datos)

print(f"Datos (n): {len(datos)}")
print(f"Número de rachas observadas (co): {co}")
print(f"Media: {mu:.4f}")
print(f"Z_0: {z:.4f}")

if abs(z) < 1.96:
    print("No se rechaza H0. Los números son independientes.")
else:
    print("Se rechaza H0. Los números no son independientes.")
```
