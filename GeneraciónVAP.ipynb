# Generación de Variables Aleatorias Poisson

La Distribución de Poisson es una distribución de probabilidad discreta que modela el número de veces que ocurre un evento en un intervalo fijo de tiempo o espacio. Nos dice cuántos eventos es probable que ocurran.

Una varaiable $X$ es Poisson con media $λ$ si:

$$p_i = P(X = i) = \frac{e^{-\lambda} \cdot \lambda^i}{i!},  i = 0,1 ,2 ,3. ...$$

Para simular esta variable usaremos la identidad:

$$p_{i+1} = \frac{\lambda}{i+1}p_i, i\geq 0$$

### Utilidad en el Mundo Real
Esta distribución es una herramienta fundamental en la ingeniería, los negocios y la ciencia porque modela situaciones donde los eventos ocurren de forma independiente y a una tasa constante. Algunos ejemplos típicos son: Llegadas de clientes, Tráfico web, Control de calidad, Salud pública, etc.

```python
import numpy as np
import random as r
import matplotlib.pyplot as plt
import math
```

```python
def poisson(n, lam):
    Lista = []
    for i in range(n):
        R = r.random()
        X = 0
        p = math.exp(-lam)   # Probabilidad individual P(X = k)
        F = p           # Probabilidad acumulada

        while R > F:
            X = X + 1
            p = (lam * p) / X
            F = F + p
        Lista.append(X)

    return Lista
```

```python
M = poisson(1000, 4)
```

```python
plt.hist(M, density=True, bins=np.arange(min(M), max(M) + 2) - 0.5)
```
