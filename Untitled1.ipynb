# Prueba de los promedios

La prueba de los promedios es una prueba estadistica que busca determinar si el conjunto de n numeros (100 en este caso) aleatorios tinen un promedio de $0.5$. se utiliza una distribuciòn normal para validar la hipotesis nula $H_0$


Las hipotesis son: //
  $H_0 = 0.5 $ //
  $H_a = 0.5 $

```python
def promedios (data, alfa):
  x = np.mean(data)
  N = len(data)
  z = (x -0.5) * np.sqrt(N)/np.sqrt(1/12)
  z_c = st.norm.ppf(1 - alfa/2)
  if abs(z) < z_c:
    print("No se rechaza la hipotesis nula: La muestra proviene de una distribución U(0, 1)")
  else:
    print("Se rechaza la hipotesis nula: La muestra no proviene de una distribución U(0, 1)")
```

```python
promedios(data, 0.05)
```
